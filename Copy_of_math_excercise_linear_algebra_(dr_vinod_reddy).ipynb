To convert an image from colour to gray scale, we sum the rgb components and rescale out of 255*3.(AXIS 1 ELEMENTS OF COLUMNS, AXIS 0 ROWS)

This returns a 2D array (matrix) of gray scale values.

```python
img = plt.imread('cat.jpg')
img_gray = img.sum(axis=-1) / (255*3)    # converting to grayscale

# create figure to plot image to
plt.figure(figsize=(4,4))
plt.imshow(img_gray, cmap='gray')
print(img_gray)
print(img_gray.shape)
print(img_gray.max())
print(img_gray.min())
```

Above we compressed the rgb vector into a scalar essentially by equally weighting the rgb channels:


 \text{gray_value} = \overrightarrow{rgb} \cdot [\frac{1}{3},\frac{1}{3},\frac{1}{3}]



However, it is common to construct the gray scale image by weighting the channels as
:
\text{gray_value} = \overrightarrow{rgb} \cdot [0.2989,0.5870,0.1140]



```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
img = plt.imread('cat.jpg')
img_gray = rgb2gray(img)

plt.figure(figsize=(4,4))
plt.imshow(img_gray, cmap="gray")
print(img_gray)
print(img_gray.shape)
print(img_gray.max())
print(img_gray.min())
```

**Separating an Image into Color Channels**

To separate an image into its colour channels we can zero out the values in the other two channels.

 This keeps the image as an array of RGB values.

```python
img = plt.imread('cat.jpg')
img_R, img_G, img_B = img.copy(), img.copy(), img.copy()
img_R[:, :, (1, 2)] = 0  # zero out GB channels
img_G[:, :, (0, 2)] = 0  # zero out RB channels
img_B[:, :, (0, 1)] = 0  # zero out RG channels
img_rgb = np.concatenate((img_R,img_G,img_B), axis=1)
plt.figure(figsize=(8, 8))
plt.imshow(img_rgb)
print(img_R.shape)
print(img_G.shape)
print(img_B.shape)
print(img_rgb.shape)
print(img)
```

Alternatively, we can split the 3D array into three 2D arrays, one for each colour channel.

Note, this is different from above where we kept the entries as RGB vectors, where just two entries were 0. This means python will still treat it as a colour image when plotting.

Below, we extract each colour channel as a matrix.

This means when plotting python will treat each colour channel matrix as grayscale.

 We print out the shapes of each matrix for emphasis.

```python
img = plt.imread('cat.jpg')
print(img.shape)
img_rch = img[:,:,0] # grab just the red channel matrix
img_gch = img[:,:,1] # grab just the green channel matrix
img_bch = img[:,:,2] # grab just the blue channel matrix
print(img_rch.shape)
print(img_gch.shape)
print(img_bch.shape)
```

STACKED IMAGE (SCALING)

```python
img_stacked = np.stack([img_rch, img_gch, img_bch], axis = 2)
plt.imshow(img_stacked)
print(img_stacked.shape)
print(img_stacked)
```

Cropping an Image

Use array slicing to crop the image.

```python
img = plt.imread('cat.jpg')

h, w, c = img.shape # height, width, channels
print(h,w)
# Calculate the center of the image
center_y, center_x = h // 2, w // 2

# Define the desired crop size
crop_height, crop_width = 200, 300

# Calculate the crop boundaries
start_y = center_y - crop_height // 2
end_y = center_y + crop_height // 2
start_x = center_x - crop_width // 2
end_x = center_x + crop_width // 2

# Ensure the crop boundaries are within the image dimensions
start_y = max(0, start_y)
end_y = min(h, end_y)
start_x = max(0, start_x)
end_x = min(w, end_x)

# Crop the image
img_crop = img[start_y:end_y, start_x:end_x, :]

print(img_crop.shape)
plt.imshow(img_crop)
plt.show()
```

**Negative of an Image**
Since colour values are between 0 and 255, we can negate an image by taking colour value m
 to 255-m
. We'll do this for both gray scale and colour images.

```python
img = plt.imread('cat.jpg')

fig = plt.figure(figsize=(10, 10))
print(fig)

# gray
img_gray = rgb2gray(img)  # convert to grayscale using function above
img_gray = 255 - img_gray
fig.add_subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Negative of Gray image')
print(img_gray)

# colour
img = 255 - img
fig.add_subplot(1, 2, 2)
plt.imshow(img)
plt.title('Negative of RGB image')
print(img)
```

**Blending (Summing) Two Images**

To blend two images we just add the arrays.

 However, we must keep in mind the RGB values are to be in the range 0 to 255, so we'll use a weighted sum.
 Also, the data type for the RGB values is uint8 so once we sum the two arrays we need to convert the entries to the right data type prior to calling imshow.

Let's blend 30% of cat.jpg with 70% of cat.jpg. First we have to make the The images have the same shape.

```python
img = plt.imread('cat.jpg')
img0 = plt.imread('cat.jpg')

# resize matrices so they have the same shape
h = min(img.shape[0],img0.shape[0])
w = min(img.shape[1],img0.shape[1])
img = img[-h:,-w:,:]  # crop image from bottom-right
img0 = img0[-h:,-w:,:] # crop image from bottom-right
print(h)
print(w)
print(img)
print(img0)

print(img.dtype)
# uint8
img_blend = (img * 0.7 + img0 * 0.3).astype(np.uint8)   # Blending them in
print(img_blend)

plt.figure(figsize=(4, 4))
plt.imshow(img_blend)
```

COLOR REDUCTION

```python
img = plt.imread('cat.jpg')

# Making Pixel values discrete by first division by // which gives int and then multiply by the same factor
img_0 = (img // 64) * 64
img_1 = (img // 128) * 128
img_all = np.concatenate((img, img_0, img_1), axis=1)
plt.figure(figsize=(8, 8))
plt.imshow(img_all)
```

BINARIZE IMAGE
A binary image is one where each rgb value is either 0 or 255.

 Below we turn an image into a binary image. We choose two different thresholds for a colour channel.

 If the value is larger than this threshold it gets bumped up to 255, if it is below this threshold it gets dropped to 0.

```python
img = plt.imread('cat.jpg')

img_64 = (img > 64) * 255
img_128 = (img > 128) * 255
fig = plt.figure(figsize=(8, 8))
img_all = np.concatenate((img, img_64, img_128), axis=1)
plt.imshow(img_all)
```
