DISCRETIZATION

```python
from sklearn.preprocessing import KBinsDiscretizer
import numpy as np
import pandas as pd

# Sample Data
data = {'Age': [10, 25, 30, 45, 55, 60, 75, 80, 90]}
df = pd.DataFrame(data)

# Reshape the data for KBinsDiscretizer (requires 2D array)
age_data = df[['Age']].values

# Equal-width discretization
discretizer_uniform = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
df['Age_SKL_Uniform'] = discretizer_uniform.fit_transform(age_data)

# Equal-frequency (quantile) discretization
discretizer_quantile = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='quantile')
df['Age_SKL_Quantile'] = discretizer_quantile.fit_transform(age_data)

print(df)
```

logarithmic/power transformation

```python
import numpy as np
import pandas as pd

# Sample data with a right-skewed distribution
data = pd.Series([1, 5, 10, 20, 50, 100, 250, 500, 1000])

# Apply natural logarithm transformation
log_transformed_data_natural = np.log(data)

# Apply base-10 logarithm transformation
log_transformed_data_base10 = np.log10(data)

# To handle zero or negative values, add a small constant before logging
data_with_zero = pd.Series([0, 1, 5, 10, 50])
log_transformed_data_plus_one = np.log1p(data_with_zero) # equivalent to np.log(data_with_zero + 1)

print("Original Data:\n", data)
print("\nNatural Log Transformed Data:\n", log_transformed_data_natural)
print("\nBase-10 Log Transformed Data:\n", log_transformed_data_base10)
print("\nLog Transformed Data (with +1 for zeros):\n", log_transformed_data_plus_one)
```

```python
from sklearn.preprocessing import PowerTransformer
import numpy as np
import pandas as pd

# Sample data
data = pd.DataFrame({'value': [1, 5, 10, 20, 50, 100, 250, 500, 1000]})

# Instantiate PowerTransformer with Box-Cox (requires positive data)
# For data that can contain zeros or negative values, use 'yeo-johnson'
pt_boxcox = PowerTransformer(method='box-cox', standardize=False)

# Fit and transform the data
boxcox_transformed_data = pt_boxcox.fit_transform(data[['value']])
data['boxcox_transformed'] = boxcox_transformed_data

# Instantiate PowerTransformer with Yeo-Johnson (handles positive, zero, and negative data)
pt_yeojohnson = PowerTransformer(method='yeo-johnson', standardize=False)

# Fit and transform the data
yeojohnson_transformed_data = pt_yeojohnson.fit_transform(data[['value']])
data['yeojohnson_transformed'] = yeojohnson_transformed_data

print("Original Data:\n", data[['value']])
print("\nBox-Cox Transformed Data:\n", data[['boxcox_transformed']])
print("\nYeo-Johnson Transformed Data:\n", data[['yeojohnson_transformed']])
```

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# Sample data
X = np.array([[1, 1], [1.1, 1.1], [1.2, 1.2], [10, 10], [1.3, 1.3]])

# Create an Isolation Forest model
model = IsolationForest(random_state=42)

# Fit the model and predict outliers (returns -1 for outliers, 1 for inliers)
predictions = model.fit_predict(X)

print(predictions)
```

```python
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

# Sample data
X = np.array([[1, 1], [1.1, 1.1], [1.2, 1.2], [10, 10], [1.3, 1.3]])

# Create a Local Outlier Factor model
# 'contamination' parameter estimates the proportion of outliers in the data
model = LocalOutlierFactor(n_neighbors=2, contamination=0.1)

# Fit the model and predict outliers (returns -1 for outliers, 1 for inliers)
predictions = model.fit_predict(X)

print(predictions)
```

```python
from sklearn.svm import OneClassSVM
import numpy as np

# Sample data
X = np.array([[1, 1], [1.1, 1.1], [1.2, 1.2], [10, 10], [1.3, 1.3]])

# Create a One-Class SVM model
# 'nu' parameter represents the upper bound on the fraction of training errors and a lower bound of the fraction of support vectors
model = OneClassSVM(gamma='auto', nu=0.1)

# Fit the model and predict outliers (returns -1 for outliers, 1 for inliers)
predictions = model.fit_predict(X)

print(predictions)
```

outlier capping

```python
    import pandas as pd
    import numpy as np

    # Sample DataFrame
    data = {'value': [10, 20, 30, 40, 50, 100, 5, 120, 150, 200, 250, 300]}
    df = pd.DataFrame(data)

    # Define lower and upper percentiles
    lower_percentile = 5
    upper_percentile = 95

    # Calculate the capping limits
    lower_limit = np.percentile(df['value'], lower_percentile)
    upper_limit = np.percentile(df['value'], upper_percentile)

    # Apply capping using np.clip
    df['value_capped'] = np.clip(df['value'], lower_limit, upper_limit)

    print(df)
```

```python
    import pandas as pd
    import numpy as np

    # Sample DataFrame
    data = {'value': [10, 20, 30, 40, 50, 100, 5, 120, 150, 200, 250, 300]}
    df = pd.DataFrame(data)

    # Calculate Q1, Q3, and IQR
    Q1 = df['value'].quantile(0.25)
    Q3 = df['value'].quantile(0.75)
    IQR = Q3 - Q1

    # Define upper and lower bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Apply capping
    df['value_capped_iqr'] = np.where(df['value'] < lower_bound, lower_bound,
                                      np.where(df['value'] > upper_bound, upper_bound, df['value']))

    print(df)
```
