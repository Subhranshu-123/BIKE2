```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame (same as above)
data = {'A': ['X', 'Y', 'X', 'Z', 'Y'],
        'B': ['P', 'Q', 'P', 'R', 'Q'],
        'C': ['M', 'N', 'M', 'O', 'N']}
df = pd.DataFrame(data)

# List of columns to plot
categorical_cols = ['A', 'B', 'C']

# Determine grid size for subplots
n_cols = len(categorical_cols)
n_rows = (n_cols + 1) // 2 # Adjust as needed for layout

fig, axes = plt.subplots(n_rows, n_cols, figsize=(5 * n_cols, 4 * n_rows))
axes = axes.flatten() # Flatten the array of axes for easy iteration

for i, col in enumerate(categorical_cols):
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Count of {col}')
    axes[i].tick_params(axis='x', rotation=45) # Rotate x-axis labels if needed

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
df = sns.load_dataset('titanic')

# Aggregate the counts for a stacked effect
# Here, we want to see 'survived' counts stacked by 'class'
stacked_data = pd.crosstab(df['class'], df['survived'])

# Plot as a stacked bar chart
stacked_data.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Survival Count by Class (Stacked)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a sample DataFrame
data = {
    'x_variable': [10, 20, 30, 40, 50],
    'y_variable': [5, 15, 25, 35, 45],
    'size_variable': [100, 500, 200, 800, 300],
    'category': ['A', 'B', 'A', 'C', 'B']
}
df = pd.DataFrame(data)

# Create the bubble chart
sns.scatterplot(
    data=df,
    x='x_variable',
    y='y_variable',
    size='size_variable',
    hue='category',
    sizes=(50, 1000),  # Control min and max bubble size
    alpha=0.7,        # Set transparency
    legend='full'     # Show the legend for size and hue
)

plt.title('Bubble Chart Example with Seaborn')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Example: Create a sample DataFrame
data = {
        'Variable_A': [10, 12, 15, 18, 20],
        'Variable_B': [5, 6, 7, 8, 9],
        'Variable_C': [2, 4, 3, 5, 6],
        'Variable_D': [8, 7, 6, 5, 4]
    }
df = pd.DataFrame(data)
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6)) # Adjust figure size as needed
sns.heatmap(
        correlation_matrix,
        annot=True,       # Display correlation coefficients on the heatmap
        cmap='coolwarm',  # Choose a colormap (e.g., 'coolwarm', 'RdBu', 'viridis')
        vmin=-1,          # Set minimum value for color scale (for correlation)
        vmax=1,           # Set maximum value for color scale (for correlation)
        center=0,         # Set the center of the color scale
        linewidths=.5,    # Add lines between cells
        linecolor='black' # Set line color
    )
plt.title('Correlation Matrix Heatmap')
plt.show()
```
