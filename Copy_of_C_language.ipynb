```python
%%writefile add_input.c
#include <stdio.h>
#include <stdlib.h> // Needed for atoi()

int main(int argc, char *argv[]) {
    // Check if the user provided 2 numbers
    if (argc < 3) {
        printf("Please provide two numbers! Example: ./add_input 10 20\n");
        return 1;
    }

    // Convert strings from command line to integers
    int num1 = atoi(argv[1]);
    int num2 = atoi(argv[2]);
    int sum = num1 + num2;

    printf("The sum of %d and %d is: %d\n", num1, num2, sum);

    return 0;
}
```

```python
%%writefile add_multiple.c
#include <stdio.h>
#include <stdlib.h>

int main(int argc, char *argv[]) {
    if (argc < 2) {
        printf("Usage: ./add_multiple <num1> <num2> ... <numN>\n");
        return 1;
    }

    int sum = 0; // The accumulator
    printf("Adding: ");

    // i starts at 1 because argv[0] is the program's name
    for (int i = 1; i < argc; i++) {
        int val = atoi(argv[i]); // Convert string argument to integer
        sum += val; // Add to our total sum

        printf("%d%s", val, (i == argc - 1) ? "" : " + ");
    }

    printf("\nTotal Sum: %d\n", sum);
    return 0;
}
```

```python
!gcc add_multiple.c -o add_multiple
```

```python
!./add_multiple 5 10 15 20 25
```

```python
%%writefile recursive_sum.c
#include <stdio.h>

// Recursive function
int sum(int n) {
    if (n <= 0) {
        return 0; // Base case: stop when n is 0
    } else {
        return n + sum(n - 1); // Recursive call: n + sum of (n-1)
    }
}

int main() {
    int number = 5;
    int result = sum(number);

    printf("The recursive sum of numbers up to %d is: %d\n", number, result);
    return 0;
}
```

```python
!gcc recursive_sum.c -o recursive_sum && ./recursive_sum
```

```python
%%writefile main.cpp
#include <iostream>
#include <string>
using namespace std;

// The Blueprint (Class)
class Dog {
public:
    string name;
    void bark() {
        cout << name << " says: Woof!" << endl;
    }
};

int main() {
    // The Instance (Object)
    Dog myDog;
    myDog.name = "Tuffy";
    myDog.bark();
    return 0;
}
```

```python
!g++ main.cpp -o main && ./main
```

```python
%%writefile main.c
#include <stdio.h>

// Struct to store data (Attributes)
struct Dog {
    char* name;
    void (*bark)(char*); // Function pointer (Method)
};

// Actual function
void dog_bark(char* name) {
    printf("%s says: Woof!\n", name);
}

int main() {
    struct Dog myDog;
    myDog.name = "Tuffy";
    myDog.bark = dog_bark; // Pointing to the function

    myDog.bark(myDog.name);
    return 0;
}
```

```python
!gcc main.c -o main && ./main
```
