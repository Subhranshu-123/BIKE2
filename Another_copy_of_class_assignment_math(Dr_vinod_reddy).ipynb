```python
import random

def simulate_coin_flips(num_flips_per_round, num_rounds):

    total_heads = 0
    total_tails = 0

    print(f"Simulating {num_rounds} rounds of {num_flips_per_round} simultaneous coin flips:\n")

    for round_num in range(1, num_rounds + 1):
        heads_in_round = 0
        tails_in_round = 0
        round_results = []

        for _ in range(num_flips_per_round):
            # 0 for tails, 1 for heads
            flip_result = random.randint(0, 1)
            if flip_result == 1:
                heads_in_round += 1
                round_results.append("Heads")
            else:
                tails_in_round += 1
                round_results.append("Tails")

        print(f"Round {round_num}: {', '.join(round_results)}")
        print(f"  Heads: {heads_in_round}, Tails: {tails_in_round}\n")

        total_heads += heads_in_round
        total_tails += tails_in_round

    print(f"--- Simulation Summary ---")
    print(f"Total Heads across all rounds: {total_heads}")
    print(f"Total Tails across all rounds: {total_tails}")
simulate_coin_flips(num_flips_per_round=10, num_rounds=1000)
```

```python
import random
import matplotlib.pyplot as plt

def simulate_coin_flips(num_coins, num_trials):

    heads_counts = []
    tails_counts = []

    for _ in range(num_trials):
        current_heads = 0
        current_tails = 0
        for _ in range(num_coins):
            # 0 for tails, 1 for heads
            if random.randint(0, 1) == 1:
                current_heads += 1
            else:
                current_tails += 1
        heads_counts.append(current_heads)
        tails_counts.append(current_tails)
    return heads_counts, tails_counts

# Parameters for the simulation
number_of_coins = 10
number_of_trials = 10000

# Run the simulation
heads, tails = simulate_coin_flips(number_of_coins, number_of_trials)

# Plotting the results
plt.figure(figsize=(12, 6))

# Plotting heads distribution
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.hist(heads, bins=range(number_of_coins + 2), align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.title(f'Distribution of Heads in {number_of_coins} Coin Flips')
plt.xlabel('Number of Heads')
plt.ylabel('Frequency')
plt.xticks(range(number_of_coins + 1))
plt.grid(axis='y', alpha=1)

# Plotting tails distribution
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.hist(tails, bins=range(number_of_coins + 2), align='left', rwidth=0.8, color='yellow', edgecolor='black')
plt.title(f'Distribution of Tails in {number_of_coins} Coin Flips')
plt.xlabel('Number of Tails')
plt.ylabel('Frequency')
plt.xticks(range(number_of_coins + 1))
plt.grid(axis='y', alpha=1)

plt.tight_layout()
plt.show()
```

```python
import random

def coin_flip_simulation(num_trials):

    heads_0_count = 0
    heads_1_count = 0
    heads_2_count = 0

    for _ in range(num_trials):
        # Simulate flipping two coins
        flip1 = random.choice([0, 1])  # 0 for tails, 1 for heads
        flip2 = random.choice([0, 1])

        heads_count = flip1 + flip2

        if heads_count == 0:
            heads_0_count += 1
        elif heads_count == 1:
            heads_1_count += 1
        elif heads_count == 2:
            heads_2_count += 1

    return heads_0_count, heads_1_count, heads_2_count

# Simulate 1000 trials
num_trials = 1000
results = coin_flip_simulation(num_trials)

# Print the results
print(f"Number of trials: {num_trials}")
print(f"Number of trials with 0 heads: {results[0]}")
print(f"Number of trials with 1 head: {results[1]}")
print(f"Number of trials with 2 heads: {results[2]}")
```
