```python
%%writefile linear_search.c
#include <stdio.h>

int main() {
    int arr[] = {10, 23, 4, 15, 7, 30};
    int target = 15;
    int n = sizeof(arr) / sizeof(arr[0]);
    int found = -1;

    for (int i = 0; i < n; i++) {
        if (arr[i] == target) {
            found = i;
            break;
        }
    }

    if (found != -1)
        printf("Element found at index: %d\n", found);
    else
        printf("Element not found.\n");

    return 0;
}
```

```python
!gcc linear_search.c -o linear_search
!./linear_search
```

```python
%%writefile algorithm.c
#include <stdio.h>

void bubbleSort(int arr[], int n) {
    for (int i = 0; i < n-1; i++)
        for (int j = 0; j < n-i-1; j++)
            if (arr[j] > arr[j+1]) {
                int temp = arr[j];
                arr[j] = arr[j+1];
                arr[j+1] = temp;
            }
}

int main() {
    int arr[] = {64, 34, 25, 12, 22, 11, 90};
    int n = sizeof(arr)/sizeof(arr[0]);
    bubbleSort(arr, n);
    printf("Sorted array: ");
    for (int i=0; i < n; i++) printf("%d ", arr[i]);
    return 0;
}
```

```python
!gcc algorithm.c -o algorithm
!./algorithm
```

```python
%%writefile quicksort.c
#include <stdio.h>

// Function to swap two elements
void swap(int* a, int* b) {
    int t = *a;
    *a = *b;
    *b = t;
}

// Partition function using the last element as pivot
int partition(int arr[], int low, int high) {
    int pivot = arr[high];
    int i = (low - 1);

    for (int j = low; j <= high - 1; j++) {
        if (arr[j] < pivot) {
            i++;
            swap(&arr[i], &arr[j]);
        }
    }
    swap(&arr[i + 1], &arr[high]);
    return (i + 1);
}

// Main Quick Sort function
void quickSort(int arr[], int low, int high) {
    if (low < high) {
        int pi = partition(arr, low, high);
        quickSort(arr, low, pi - 1);
        quickSort(arr, pi + 1, high);
    }
}

int main() {
    int data[] = {10, 7, 8, 9, 1, 5};
    int n = sizeof(data) / sizeof(data[0]);

    quickSort(data, 0, n - 1);

    printf("Sorted array: ");
    for (int i = 0; i < n; i++) printf("%d ", data[i]);
    return 0;
}
```

```python
%%writefile bloom.c
#include <stdio.h>
#include <stdbool.h>
#include <string.h>

#define SIZE 100 // Size of the bit array

bool bitArray[SIZE] = {0};

// Simple hash function 1
int hash1(char *str) {
    unsigned int hash = 5381;
    int c;
    while ((c = *str++)) hash = ((hash << 5) + hash) + c;
    return hash % SIZE;
}

// Simple hash function 2
int hash2(char *str) {
    unsigned int hash = 0;
    int c;
    while ((c = *str++)) hash = c + (hash << 6) + (hash << 16) - hash;
    return hash % SIZE;
}

void insert(char *str) {
    bitArray[hash1(str)] = 1;
    bitArray[hash2(str)] = 1;
}

bool lookup(char *str) {
    if (bitArray[hash1(str)] && bitArray[hash2(str)])
        return true; // Probably present
    return false;    // Definitely not present
}

int main() {
    insert("google");
    insert("colab");

    printf("'google': %s\n", lookup("google") ? "Probably Yes" : "No");
    printf("'colab': %s\n", lookup("colab") ? "Probably Yes" : "No");
    printf("'algorithm': %s\n", lookup("algorithm") ? "Probably Yes" : "No");

    return 0;
}
```

```python
%%writefile trading_filter.c
#include <stdio.h>

void calculate_sma(float prices[], int n, int period) {
    printf("Day\tPrice\tSMA(%d)\n", period);
    for (int i = 0; i < n; i++) {
        if (i < period - 1) {
            printf("%d\t%.2f\tN/A\n", i + 1, prices[i]);
        } else {
            float sum = 0;
            for (int j = 0; j < period; j++) {
                sum += prices[i - j];
            }
            printf("%d\t%.2f\t%.2f\n", i + 1, prices[i], sum / period);
        }
    }
}

int main() {
    float prices[] = {150.20, 152.10, 151.05, 153.50, 155.00, 154.20, 156.80};
    int n = sizeof(prices) / sizeof(prices[0]);
    calculate_sma(prices, n, 3); // 3-day SMA filter
    return 0;
}
```

```python
!gcc trading_filter.c -o trading_filter
!./trading_filter
```
