```python
import plotly.graph_objects as go
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a scatter trace
scatter_trace = go.Scatter(x=x, y=y, mode='markers')

# Create a figure and add the trace
fig = go.Figure(data=[scatter_trace])

# Customize the layout (e.g., title)
fig.update_layout(title='Simple Scatter Plot')

# Display the plot
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# 1. Create Sample Data (or load your own DataFrame)
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'A', 'B', 'C'],
    'Value': [1, 2, 3, 4, 5, 6, 7, 8, 9, 1.5, 5.5, 8.5]
}
df = pd.DataFrame(data)

# 2. Create the Violin Plot
fig = px.violin(
    df,
    x="Category",  # Categorical variable for x-axis
    y="Value",      # Continuous variable for y-axis
    box=True,      # Overlay a box plot inside the violin
    points="all",   # Display individual data points
    color="Category", # Group and color violins by category
    title="Violin Plot of Value by Category", # Add a title to the plot
    labels={'Category': 'Product Category', 'Value': 'Measured Value'}, # Customize axis labels
    violinmode='group' #  Draw violins side-by-side (default)
    # You can also use 'overlay' to draw violins on top of each other
)

# 3. Display the plot
fig.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px

# Load sample data (Plotly's built-in iris dataset)
df = px.data.iris()

# Create a box plot trace
box_trace = go.Box(
    x=df['species'],  # Categorical variable for x-axis
    y=df['sepal_length'], # Numerical variable for y-axis
    name='Sepal Length',
    boxpoints='all',  # Display all data points
    notched=True,     # Create a notched box plot
    marker=dict( # Customize marker style
        color='purple',
        symbol='star',
        size=6
    ),
    line=dict(width=2) # Customize line style
)

# Create figure and add trace
fig = go.Figure(box_trace)

# Update layout (title and axis labels)
fig.update_layout(title='Box Plot of Sepal Length by Species', # Set plot title
                  xaxis_title='Species',  # Set x-axis title
                  yaxis_title='Sepal Length (cm)') # Set y-axis title

fig.show()
```

```python
import plotly.express as px
import pandas as pd

schools = ["Brown", "NYU", "Notre Dame", "Cornell", "Tufts", "Yale"]
n_schools = len(schools)
women_salary = [72, 67, 73, 80, 76, 79]
men_salary = [92, 94, 100, 107, 112, 114]

df = pd.DataFrame(dict(school=schools*2, salary=men_salary + women_salary, gender=["Men"]*n_schools + ["Women"]*n_schools))

fig = px.scatter(df, x="salary", y="school", color="gender",
                 title="Gender Earnings Disparity",
                 labels={"salary":"Annual Salary (in thousands)"})
fig.show()
```

```python
import plotly.express as px
import pandas as pd

dates = pd.date_range(start="2024-01-01", periods=10, freq='D')
data = {
    "Date": dates.tolist() * 2,
    "Sales": [100, 120, 150, 130, 170, 180, 200, 190, 210, 230] + [80, 110, 140, 120, 160, 170, 190, 180, 200, 220],
    "Category": ["A"] * 10 + ["B"] * 10
}
df = pd.DataFrame(data)

fig = px.area(df, x="Date", y="Sales", color="Category", title="Stacked Area Chart")
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Sample Data
data = {
    'Region': ['North', 'North', 'South', 'South', 'North', 'South'],
    'Country': ['USA', 'Canada', 'Brazil', 'Argentina', 'USA', 'Brazil'],
    'Sales': [100, 75, 120, 90, 50, 60]
}
df = pd.DataFrame(data)

# Create Sunburst Chart
fig = px.sunburst(df, path=['Region', 'Country'], values='Sales',
                  title='Sales Distribution by Region and Country',
                  color='Region',
                  color_discrete_map={'North':'lightblue', 'South':'lightcoral'})

# Display the chart
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Generate sample data for x, y, and z axes
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2)) # Example: a 3D surface

# Create a 3D surface trace
surface_trace = go.Surface(z=Z, x=X, y=Y)

# Create a figure and add the trace
fig = go.Figure(data=[surface_trace])

# Update layout for title and axis labels
fig.update_layout(
    title='3D Surface Plot Example',
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    )
)

# Display the figure
fig.show()
```
