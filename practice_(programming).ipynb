```python
def clean_strings(strings):
  result=[]
  for value in strings:
    value = value.strip()
    value = re.sub("[!#?]", "", value)
    value = value.title()
    result.append(value)
  return result

cleaned_states = clean_strings(states)
print(cleaned_states)
```

**Example 2: Filter for numbers greater than 5**

If you want to get all numbers greater than 5, you can use a lambda function that checks this condition:

```python
numbers = [4, 0, 8, 6, 10, 1, 3, 7, 5]
greater_than_five = list(filter(lambda x: x > 5, numbers))
print(greater_than_five)
```

### Consuming a Generator

**1. Using a `for` loop:**

The most common way to consume a generator is to iterate over it with a `for` loop:

```python
gen = _make_gen()
for value in gen:
    print(value)
```

**2. Using the `next()` function:**

You can also get the next value from a generator by using the `next()` function. This will raise a `StopIteration` exception when the generator is exhausted.

```python
gen = _make_gen()
print(next(gen))
print(next(gen))
```

**3. Converting to a list:**

If you want all the values from the generator at once, you can convert it to a list. Be careful with this approach, as it can consume a lot of memory if the generator produces a large number of values.

```python
gen = _make_gen()
all_values = list(gen)
print(all_values)
```
