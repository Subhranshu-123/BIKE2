Check Stationarity

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(ts['Original Principal Amount'])
print(result)
```

Stationary means:

Mean = constant

Differencing (if needed)

```python
ts_diff = ts.diff().dropna()
print(ts_diff)
```

Removes trend

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(ts_diff['Original Principal Amount'])
plot_pacf(ts_diff['Original Principal Amount'])
plt.show()
```

Build ARIMA Model

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(ts['Original Principal Amount'], order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
plt.plot(ts, label='Original')
plt.plot(forecast, label='Forecast')
plt.legend()
plt.show()
```

Convert to Monthly Data

```python
ts_monthly = df.resample('M').sum()
print(ts_monthly)
```

Seasonality is usually monthly / quarterly
Yearly data cannot show seasonal patterns

Calculate Moving Average (Trend)

```python
ts_monthly['MA'] = ts_monthly['Original Principal Amount'].rolling(window=12).mean()
print(ts_monthly['MA'])
```

Removes seasonality
Extracts trend component (T)

Compute Seasonal Ratio

```python
ts_monthly['Seasonal Ratio'] = ts_monthly['Original Principal Amount'] / ts_monthly['MA']
print(ts_monthly['Seasonal Ratio'])
```

S=T/Y
This isolates seasonality

 Extract Month

Needed to group values by month (Jan, Feb, etc.)

```python
ts_monthly['Month'] = ts_monthly.index.month
print(ts_monthly['Month'])
```

```python
seasonal_index = ts_monthly.groupby('Month')['Seasonal Ratio'].mean()
print(seasonal_index)
```

Calculate Seasonal Index

```python
seasonal_index = ts_monthly.groupby('Month')['Seasonal Ratio'].mean()
print(seasonal_index)
```

Average seasonal effect for each month
Gives final seasonal index

Normalize Seasonal Index

```python
seasonal_index = seasonal_index / seasonal_index.mean()
print(seasonal_index)
```

Ensures average = 1

Interpretation
Value	Meaning
> 1	High season
< 1	Low season
= 1	Normal

```python
seasonal_index.plot(kind='bar', title='Seasonal Index')
plt.show()
```

Easy visualization of peak and low months

A power transformation technique used to:

Stabilize variance
Make data more normal (Gaussian)
Improve forecasting models

```python
from scipy.stats import boxcox

# Ensure no zero or negative values
ts_positive = ts['Original Principal Amount'][ts['Original Principal Amount'] > 0]

ts['BoxCox'], lam = boxcox(ts_positive)

print("Lambda value:", lam)
```

Why this step?
Box-Cox requires:
Positive values only
It returns:
Transformed data
Optimal λ (lambda)

Compare Transformations

```python
plt.figure(figsize=(10,5))

plt.plot(ts['Original Principal Amount'], label='Original')
plt.plot(ts['BoxCox'], label='Box-Cox')

plt.legend()
plt.title("Original vs Box-Cox Transformation")
plt.show()
```

Why?
Shows:
Smoother curve
Reduced variance
Better stability

Check Normality

```python
import scipy.stats as stats

stats.probplot(ts['BoxCox'].dropna(), dist="norm", plot=plt)
plt.title("Q-Q Plot for Box-Cox")
plt.show()
```

Revenue Formula:
Revenue=min⁡(Demand,Capacity)×Price
Revenue=min(Demand,Capacity)×Price

If demand > capacity → you can only sell up to capacity
If demand < capacity → revenue depends on demand

Forecast Demand

```python
predicted_demand = model_fit.forecast(steps=30)
print(predicted_demand)
```

Revenue depends on future demand
Forecast gives expected values

Define Business Parameters

```python
capacity = 50000000   # total loan capacity (example)
price = 0.08          # interest rate (8%)
```

Capacity → max loan disbursement possible
Price → interest rate (acts like ticket price)

Compute Revenue

```python
revenue = []

for demand in predicted_demand:
    rev = min(demand, capacity) * price
    revenue.append(rev)
    print(f"Demand: {demand}, Revenue: {rev}")

total_revenue = sum(revenue)
```

Calculates revenue for each period

Convert to DataFrame

```python
rev_df = pd.DataFrame({
    'Demand': predicted_demand,
    'Revenue': revenue
})
print(rev_df)
```

Plot Revenue

```python
import matplotlib.pyplot as plt

plt.plot(rev_df['Revenue'])
plt.title("Revenue Forecast")
plt.show()
```

Visual understanding of:

Growth
Decline
Stability
