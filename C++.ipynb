```python
%%writefile calculator.cpp
#include <iostream>
using namespace std;

class Calculator {
public:
    // Method to add two numbers
    double add(double a, double b) {
        return a + b;
    }

    // Method to subtract two numbers
    double subtract(double a, double b) {
        return a - b;
    }
};

int main() {
    // Create an object of the Calculator class
    Calculator myCalc;

    double n1 = 50.5, n2 = 25.3;

    // Use the object's methods
    cout << "Addition: " << myCalc.add(n1, n2) << endl;
    cout << "Subtraction: " << myCalc.subtract(n1, n2) << endl;

    return 0;
}
```

```python
!g++ calculator.cpp -o calculator && ./calculator

```
