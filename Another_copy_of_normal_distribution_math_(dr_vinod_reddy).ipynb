```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mean = 171.380750
std_dev = 11.077529


x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 500)

# Calculate the Probability Density Function (PDF) for the normal distribution
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Plot the normal distribution curve
plt.plot(x, pdf, label=f'Normal Distribution (Î¼={mean}, Ïƒ={std_dev})')

# Plot the mean as a vertical line
plt.axvline(mean, color='red', linestyle='--', label=f'Mean (Î¼={mean})')

# Plot lines for one standard deviation above and below the mean
plt.axvline(mean + std_dev, color='green', linestyle=':', label=f'Î¼ + Ïƒ ({mean + std_dev:.2f})')
plt.axvline(mean - std_dev, color='green', linestyle=':', label=f'Î¼ - Ïƒ ({mean - std_dev:.2f})')

# Add labels, title, and legend
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution with Mean and Standard Deviation')
plt.legend()
plt.grid(True)
plt.show()
```
