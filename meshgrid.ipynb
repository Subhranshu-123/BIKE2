```python
# Define the coordinate vectors
x = np.arange(-5, 5.1, 0.5) # Using arange for range and step
y = np.arange(-5, 5.1, 0.5)
z = np.arange(-5, 5.1, 0.5)

# Create the 3-D grid
X, Y, Z = np.meshgrid(x, y, z)

# Calculate the distance from the origin (R)
R = np.sqrt(X**2 + Y**2 + Z**2) # Using ** for power and np.sqrt for square root

# Calculate the function values
# Handle division by zero by replacing 0 with a small number (e.g., 1e-9)
V = np.sin(R) / (R + 1e-9)

# You can then visualize this data using functions like slice, volumerender, etc.
# For example, to visualize a slice:
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # Create a 3D subplot

# To visualize slices, we can use contourf on slices of the 3D data
# Here's an example showing a slice at z=0
z_slice_index = np.argmin(np.abs(z - 0)) # Find the index closest to z=0
ax.contourf(X[:,:,z_slice_index], Y[:,:,z_slice_index], V[:,:,z_slice_index], cmap='viridis')


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Slice at Z=0')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define 1D arrays for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create the 2D coordinate grids
X, Y = np.meshgrid(x, y)

# Define a function of two variables
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')
plt.show()
```
