```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions

# Load the Iris dataset (a popular dataset for classification)
iris = datasets.load_iris()
X = iris.data[:, [0, 2]]  # Select the first two features for 2D plotting
y = iris.target

# Train a Support Vector Machine (SVM) classifier
svm = SVC(C=0.5, kernel='linear')
svm.fit(X, y)

# Plot the decision regions
plot_decision_regions(X, y, clf=svm, legend=2)

# Add labels and title
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris')
plt.show()
```

```python
from mlxtend.plotting import plot_linear_regression
import matplotlib.pyplot as plt
import numpy as np

# Example data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 6]).reshape(-1,1)

# Plot the linear regression fit
intercept, slope, corr_coeff = plot_linear_regression(X, y)

# Display the plot
plt.show()

print(f"Intercept: {intercept:.2f}")
print(f"Slope: {slope:.2f}")
print(f"Correlation Coefficient: {corr_coeff:.2f}")
```

```python
from mlxtend.data import iris_data
from mlxtend.plotting import ecdf
import matplotlib.pyplot as plt

# Load sample data
X, y = iris_data()
x1 = X[:, 0]  # Select the first feature

# Create the ECDF plot
ax, _, _ = ecdf(x1)

# Add labels and title (optional)
ax.set_xlabel('Feature Value')
ax.set_ylabel('ECDF')
ax.set_title('Empirical Cumulative Distribution Function')

# Display the plot
plt.show()
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
knn = KNeighborsClassifier(n_neighbors=4)

sfs = SFS(knn,
          k_features=4,
          forward=True,
          floating=False,
          scoring='accuracy',
          cv=5)

sfs = sfs.fit(X, y)

fig1 = plot_sfs(sfs.get_metric_dict(),
                kind='std_dev',
                figsize=(6, 4))

plt.ylim([0.8, 1])
plt.title('Sequential Forward Selection (w. StdDev)')
plt.grid()
```
