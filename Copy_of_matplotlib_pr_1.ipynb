```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
sub_category1_values = [10, 20, 15, 25]
sub_category2_values = [5, 12, 8, 10]
sub_category3_values = [7, 8, 10, 5]

# Plot the first set of bars
plt.bar(categories, sub_category1_values, label='Sub-Category 1')

# Plot the second set of bars on top of the first
# The 'bottom' argument is crucial for stacking
plt.bar(categories, sub_category2_values, bottom=sub_category1_values, label='Sub-Category 2')

# Plot the third set of bars on top of the cumulative sum of the first two
bottom_cumulative = np.array(sub_category1_values) + np.array(sub_category2_values)
plt.bar(categories, sub_category3_values, bottom=bottom_cumulative, label='Sub-Category 3')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
plt.legend() # Display the legend for sub-categories
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
group1_values = [10, 15, 7, 12]
group2_values = [8, 18, 10, 9]

bar_width = 0.35
x = np.arange(len(categories))

plt.bar(x - bar_width/2, group1_values, bar_width, label='Group 1', color='skyblue')
plt.bar(x + bar_width/2, group2_values, bar_width, label='Group 2', color='lightcoral')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart Example')
plt.xticks(x, categories) # Set x-axis tick positions and labels
plt.legend()
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories = ['Category A', 'Category B', 'Category C', 'Category D']
group1_values = [10, 25, 15, 30]
group2_values = [12, 20, 18, 25]
group3_values = [8, 22, 10, 28]
bar_width = 0.25
r1 = np.arange(len(categories))

r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
plt.bar(r1, group1_values, color='skyblue', width=bar_width, label='Group 1')
plt.bar(r2, group2_values, color='lightcoral', width=bar_width, label='Group 2')
plt.bar(r3, group3_values, color='lightgreen', width=bar_width, label='Group 3')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Multiple Bar Chart Example')
plt.xticks([r + bar_width for r in range(len(categories))], categories) # Center x-ticks
plt.legend()
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data Generation
np.random.seed(42)  # For reproducibility
df = pd.DataFrame(np.random.rand(50, 4), columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])

num_features = df.shape[1]
fig, axes = plt.subplots(num_features, num_features, figsize=(10, 10))

for i in range(num_features):
    for j in range(num_features):
        ax = axes[i, j]
        if i == j:
            # Histogram on the diagonal
            ax.hist(df.iloc[:, i], bins=15, color='skyblue', edgecolor='black')
        else:
            # Scatter plot elsewhere
            ax.scatter(df.iloc[:, j], df.iloc[:, i], alpha=0.7, s=10, color='blue')

        # Formatting
        if j == 0:
            ax.set_ylabel(df.columns[i]) # Labels on leftmost column
        else:
            ax.set_yticklabels([])
        if i == num_features - 1:
            ax.set_xlabel(df.columns[j]) # Labels on bottom row
        else:
            ax.set_xticklabels([])

        ax.tick_params(left=False, right=False, labelleft=False,
                       labelbottom=False, bottom=False)  # Remove ticks for a cleaner look


plt.tight_layout() # Prevent labels from overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

data = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5])

# Count the frequency of each unique value
value_counts = Counter(data)

fig, ax = plt.subplots()

for value, count in value_counts.items():
    # Plot dots for each frequency
    for i in range(count):
        ax.plot(value, i, 'o', markersize=8) # 'o' for circle marker

ax.set_xlabel("Value")
ax.set_ylabel("Frequency")
ax.set_title("Frequency Dot Plot")
ax.set_xticks(sorted(value_counts.keys())) # Ensure all values are shown on x-axis
ax.set_yticks(range(max(value_counts.values()) + 1)) # Set y-axis ticks based on max frequency
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Sample hierarchical data
data = {
    'parent': ['', 'A', 'A', 'B', 'B', 'C', 'C'],
    'child': ['Root', 'A1', 'A2', 'B1', 'B2', 'C1', 'C2'],
    'value': [0, 10, 15, 8, 12, 5, 7]
}
df = pd.DataFrame(data)

# Create the sunburst chart
fig = px.sunburst(
    df,
    names='child',
    parents='parent',
    values='value',
    title='Sample Sunburst Chart'
)
fig.show()
```
