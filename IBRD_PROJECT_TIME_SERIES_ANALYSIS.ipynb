Box-Cox Transformation — “Advanced Stabilization”

When log is not enough, Box-Cox is used. It automatically finds the best transformation to make data more “normal”

```python
import numpy as np
from scipy import stats

# Add a small constant to ensure all values are positive for Box-Cox transformation
df['Original Principal Amount_positive'] = df['Original Principal Amount'] + 1e-6

# Apply transformation
# 'xt' is the transformed data, 'lmbda' is the optimal lambda found
xt, lmbda = stats.boxcox(df['Original Principal Amount_positive'])
print(f"Optimal Lambda: {lmbda}")
```



ARIMA Model — “Combining Everything”


Now we combine everything:


p → from PACF d → number of differencing q → from ACF

We train the model using past data


```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['Original Principal Amount'], order=(1,1,1))
model_fit = model.fit()

print(model_fit.summary())
```

Forecasting — “Seeing the Future

```python
forecast = model_fit.forecast(steps=10)

print(forecast)
```

Plot Forecast vs Actual

```python
plt.plot(df['Original Principal Amount'], label='Actual')
plt.plot(forecast, label='Forecast', color='red')
plt.legend()
plt.show()
```

```python
auto_model = auto_arima(
    df['Original Principal Amount'],
    seasonal=False,        # Set True if seasonal data
    stepwise=True,
    trace=True,            # Shows model selection process
    suppress_warnings=True
)

print(auto_model.summary())
```

```python
forecast_auto = auto_model.predict(n_periods=10)

print(forecast_auto)
```

Compare Manual vs Auto ARIMA

```python
plt.plot(df['Original Principal Amount'], label='Actual')
plt.plot(forecast, label='Manual ARIMA')
plt.plot(forecast_auto, label='Auto ARIMA')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

def evaluate_model(actual, predicted):
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    return rmse, mape
```



Model Evaluation — “How Good Are Our Predictions?”

RMSE → penalizes large errors MAPE → gives % error (business-friendly)

Lower RMSE = better model Sensitive to large errors


```python
# Example: align actual values with forecast horizon
actual_arima = df['Original Principal Amount'][-len(forecast):].values
rmse_arima, mape_arima = evaluate_model(actual_arima, forecast.values)

actual_auto = df['Original Principal Amount'][-len(forecast_auto):].values
rmse_auto, mape_auto = evaluate_model(actual_auto, forecast_auto.values)
print(rmse_arima, mape_arima)
print(rmse_auto, mape_auto)
```

```python
import pandas as pd

results = pd.DataFrame({
    "Model": ["Manual ARIMA", "Auto ARIMA"],
    "RMSE": [rmse_arima, rmse_auto],
    "MAPE (%)": [mape_arima, mape_auto]
})

results.sort_values(by="RMSE", inplace=True)
print(results)
```
